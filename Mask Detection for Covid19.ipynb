```python
from termcolor import colored
import cv2
face_cascade = cv2.CascadeClassifier(r'C:\Users\sonia\Downloads\FaceSmileDetection-master\haarcascade_frontalface_default.xml')
smile_cascade = cv2.CascadeClassifier(r'C:\Users\sonia\Downloads\FaceSmileDetection-master\haarcascade_smile.xml')

def detect(gray, frame):
    faces = face_cascade.detectMultiScale(gray,1.3,5)
    for (x,y,w,h) in faces:
        cv2.rectangle(frame, (x,y), (x+w,y+h), (255,0,0), 2)#thickness
        roi_gray = gray[y:y+h, x:x+w]#region of interest
        roi_color = frame[y:y+h, x:x+w]
        smiles = smile_cascade.detectMultiScale(roi_gray,1.7, 25)
        p = "Mask weared"
        for (sx,sy,sw,sh) in smiles:
            cv2.rectangle(roi_color,(sx,sy),(sx+sw,sy+sh),(0,0,255),2)
            p = "Mask not weared"
        if p == "Mask not weared":
            print(colored(p,"red"))
        if p == "Mask weared":
            print(colored(p,"green"))
    return frame

video_capture = cv2.VideoCapture(0)
while True:
    _, frame = video_capture.read()
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    canvas = detect(gray, frame)
    cv2.imshow('video',canvas)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
video_capture.release()
cv2.destroyAllWindows()
```
